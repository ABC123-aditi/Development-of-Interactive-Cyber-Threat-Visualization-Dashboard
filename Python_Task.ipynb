1. **Variable**:


A variable is a name used to store data in a program.
The value of a variable can change while the program is running.

```python

name = "Alice"
age = 25
height = 5.4

print("Name:", name)
print("Age:", age)
print("Height:", height)
```

2. **Datatype**:

A data type tells Python what kind of value a variable is storing. Different data types are used to store different kinds of data like numbers, text, or true/false values.
 Here are some of the most common ones:


 **Integer(int)** :Used to store whole numbers.

  **Float(float)**:Used to store decimal numbers.

  **String(str)**:Used to store text or characters. Strings are written inside quotes.

 **Boolean(bool)**:Used to store only True or False values.

```python
x = 10
y = 5.5
z = "Hello"
a = False

print(x, type(x))
print(y, type(y))
print(z, type(z))
print(a, type(a))
```

**3.Lists**

A list is a data type used to store multiple values in a single variable.
Lists are ordered, mutable (can be changed), and written using square brackets[]. A list can store different types of data like numbers, strings, or booleans.

```python
fruits = ["Apple", "Banana", "Mango"]
fruits.append("guava")
print(fruits)
print(type(fruits))

```

**4.Tuples**

A tuple is a data type used to store multiple values in a single variable.
Tuples are ordered and immutable, which means their values cannot be changed after creation.
Tuples are written using round brackets ( ).

```python
# Creating a tuple
colors = ("Red", "Green", "Blue","Red")
print(colors)
print(type(colors))

colors.index("Blue")
```

**5.Sets**:

A set is an unordered collection of unique items. Sets are defined by enclosing elements in curly braces {} (or using the set() constructor for an empty set). Duplicate elements are automatically removed.

```python
numbers = {1, 2, 3, 4, 4}

print(numbers)
print(type(numbers))
```

**6.Dictionary**

A dictionary is a data type used to store data in keyâ€“value pairs.
Each value is accessed using its key.
Dictionaries are unordered, mutable, and written using curly brackets { }.

```python
student = {
    "name": "Alice",
    "age": 20,
    "course": "Python"
}

print(student)
print(type(student))
```

**Basic Plots with Matplotlib**

Let's use the data_df DataFrame we created earlier to demonstrate some plots.

**Line Plot**

Useful for showing trends over time or ordered categories. Here, we'll plot Salary against Experience (assuming Experience implies an ordering).

```python
plt.figure(figsize=(12, 5))
plt.plot(data_df['Experience'], data_df['Salary'], marker='o', linestyle='-')
plt.title('Salary vs. Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.grid(True)
plt.show()
```

**Scatter Plot**

Ideal for showing the relationship between two numerical variables.

Used for comparing quantities of different categories. Let's compare the average salary by city.

**Object-Oriented Programming (OOP) Concepts in Python**

**OOPS (Object-Oriented Programming System)** is a programming approach that organizes code using objects and classes.
It helps in code reusability, security, and easy maintenance.

**1.Class**

A class is a blueprint or template used to create objects.

Purpose:

*   **Blueprint:** It provides a way to create many similar objects without writing the same code repeatedly.
* **Organization:** It helps organize code into logical, manageable units.

*   **Reusability**: Code defined in a class can be reused many times to create different objects.


```python
class Student:
    name = "Aditi"
    age = 20
```

**2. Objects**

 An object (also called an instance) is a specific realization of a class. It's a concrete entity created from the class blueprint, having its own unique set of data (attributes) and capabilities (methods).

Purpose:

**Data Representation:** Objects represent real-world entities or abstract concepts in your program, holding specific data.

**Action Execution:** They can perform actions defined by their class's methods.

**Modularity:** Programs are built by interacting with various objects.

```python
s1 = Student()
print(s1.name)
print(s1.age)
```

**3. Inheritance**

Inheritance is a mechanism that allows a new class (subclass or child class) to inherit attributes and methods from an existing class (superclass or parent class). This promotes code reusability and establishes a natural hierarchy between classes.

Purpose:

**Code Reusability:** Child classes can reuse code defined in parent classes without rewriting it.

**Hierarchy:** It models an "is-a" relationship (e.g., a Cat is a Animal).

**Extension:** Child classes can add new attributes and methods or override existing ones to specialize their behavior.

```python
class Animal:
    def sound(self):
        print("Animal makes a sound")

class Dog(Animal):
    def bark(self):
        print("Dog barks")

d = Dog()
d.sound()
d.bark()
```

**4. Encapsulation:**

 Encapsulation is the bundling of data (attributes) and the methods (functions) that operate on the data into a single unit (a class). It also involves restricting direct access to some of an object's components, meaning that internal data of an object is hidden from the outside world and only accessible through predefined methods.

Purpose:

**Data Protection:** Prevents direct, unauthorized modification of an object's internal state.

**Maintainability:** Changes to internal data structures or logic within a class don't affect other parts of the code as long as the public methods remain consistent.

**Simplicity:** Users of a class only need to know how to use its public methods, not its internal workings.
Syntax (in Python, encapsulation is by convention): Python doesn't have strict private keywords like some other languages. Instead, it uses conventions:

**Public members:** Normal attributes/methods (e.g., name, get_info). Accessible from anywhere.

**Protected members:** Prefixed with a single underscore (e.g., _protected_attribute). Indicates to developers that this is for internal use or by subclasses, but it's still accessible externally.

**Private members:** Prefixed with double underscores (e.g., __private_attribute). Python performs name mangling (e.g., _ClassName__private_attribute) to make them harder to access directly from outside the class, but they are not truly private.

```python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance

    def get_balance(self):
        return self.__balance
```

**5. Polymorphism**

 Polymorphism (meaning "many forms") allows objects of different classes to be treated as objects of a common superclass. It enables a single interface to represent different underlying forms or data types. In Python, this is often achieved through method overriding and duck typing.

Purpose:

**Flexibility and Extensibility:** New classes can be added with minimal changes to existing code, as long as they adhere to the common interface.

**Code Reusability:** A single piece of code (e.g., a function) can operate on objects of various types, as long as those objects have the required methods.

**Simplified Interfaces:** It allows you to design generic interfaces for related objects, making your code cleaner and easier to understand.

```python
class Bird:
    def fly(self):
        print("Bird can fly")

class Penguin(Bird):
    def fly(self):
        print("Penguin cannot fly")

b = Bird()
p = Penguin()

b.fly()
p.fly()
```

**6. Abstraction**

Abstraction focuses on showing only essential information and hiding the complex implementation details. It allows you to create a general idea or blueprint of a system without specifying all the low-level details. In Python, abstraction can be achieved using abstract classes and abstract methods, often with the abc (Abstract Base Classes) module.

Purpose:

**Complexity Management:** It helps manage complexity by presenting a simplified view of functionality to the user, who doesn't need to know how it works internally.

**Focus on "What" not "How":** It allows designers to focus on what an object does rather than how it achieves its functionality.

**Enforce Structure:** Abstract classes can enforce that subclasses implement certain methods, ensuring a consistent interface across related classes.

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):
    @abstractmethod
    def start(self):
        pass

class Car(Vehicle):
    def start(self):
        print("Car starts with key")

c = Car()
c.start()
```

**7. Methods**

Methods are functions defined inside a class that perform actions on the object's data. They can be:

**Instance Methods:** Operate on the instance's attributes (e.g., bark(), description()). They always take self as their first parameter.

**Class Methods:** Operate on class attributes and are defined using the @classmethod decorator. They take cls (the class itself) as their first parameter.

**Static Methods:** Do not operate on instance or class attributes and are defined using the @staticmethod decorator. They don't take self or cls.

```python
try:
    numerator = 10
    denominator = 0
    result = numerator / denominator
    print(f"Result: {result}") # This line won't be reached
except ZeroDivisionError:
    print("Error: Cannot divide by zero!")

print("Program continues after handling the error.")
```
